```python
# PROPER viewing of DF

import os

cmd = "zgrep '^#' " + vcf_file + "|tail -n 1"
```

```python
columns = os.popen(cmd).read()
columns
```

```python
# Removing characters

columns = columns.strip('#').strip('\n').split('\t')
```

```python
columns
```

```python
df_variant = pd.read_csv(vcf_file, sep='\t' , comment='#', names = columns)

df_variant.shape
df_variant.head()
```
